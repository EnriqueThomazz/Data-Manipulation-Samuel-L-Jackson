# Gold Transformations Samuel L. Jackson movies and series

```python
from pyspark.sql import SparkSession
from IPython.display import display as disp

spark = SparkSession.builder.getOrCreate()
```

### Load data

```python
movies = spark.read.parquet("../Silver/movies/*")

series = spark.read.parquet("../Silver/series/*")

genres = spark.read.parquet("../Silver/genres/*")
```

```python
movies.toPandas()
```

```python
genres.toPandas()
```

```python
# Getting only the relevant columns

# He's never done adult movies
# backdrop_path, credit_id, order, poster_path and video are useless for this analysis
movies_df = movies.drop("adult", "backdrop_path", "credit_id", "order", "poster_path", "video")

series_df = series.drop("adult", "backdrop_path", "credit_id", "poster_path")
```

```python
movies_df.toPandas()
```

```python
# There are some series in wich samuel did himself. Those are listed as "Self" or "Samuel L. Jackson". I'll padronize that.

series_df.filter("character == 'Self' or character == 'Samuel L. Jackson'").toPandas()
```

```python
from pyspark.sql.functions import when

series_df = series_df.withColumn("character", when(series_df.character == "Samuel L. Jackson", "Self").otherwise(series.character))

series_df.toPandas()
```

### Creating dimensions

#### Date dimension

```python
import pyspark.sql.functions as f
from pyspark.sql.functions import monotonically_increasing_id 

# Getting all dates from movies and series
date_dim = movies_df.withColumnRenamed("release_date", "date").select("date")

date_dim = date_dim.union(series_df.withColumnRenamed("first_air_date", "date").select("date"))

# Removing duplicated dates
date_dim = date_dim.drop_duplicates()

# Adding and index
date_dim = date_dim.select("*").withColumn("date_id", monotonically_increasing_id())

# Extracting relevant information from date

# Day, Month and Year
date_dim = date_dim.withColumns({"date_day": f.dayofmonth(date_dim.date), "date_month": f.month(date_dim.date), "date_year": f.year(date_dim.date)})

# Month Name

date_dim = date_dim.withColumn('date_month_name', f.date_format('date', 'MMMM'))

# Weekday
date_dim = date_dim.withColumn('date_weekday', f.dayofweek('date'))

# Name of the week day
date_dim = date_dim.withColumn('date_weekday_name', f.date_format('date', 'EEEE'))

# Quarter
date_dim = date_dim.withColumn('date_quarter', f.quarter('date'))

date_dim.toPandas()
```

#### Genre dimension

```python
genre_dim = genres

genre_dim.toPandas()
```

#### Media dimension

```python
# The ideia of a media dimension is to merge participations in series and movies, so that the fact table only points at one dimension for audiovisual media

# Using SQL because its easier...
movies_df.createOrReplaceTempView("movies_df")
series_df.createOrReplaceTempView("series_df")

media_dim = spark.sql(
"""SELECT id as media_id, 'Series' as media_type, original_language as media_original_language, name as media_title, name as media_original_title, job as media_job, department as media_department, character as media_character, overview as media_overview
FROM series_df

UNION

SELECT id as media_id, 'Movie' as media_type, original_language as media_original_language, title as media_title, original_title as media_original_title, job as media_job, department as media_department, character as media_character, overview as media_overview
FROM movies_df""")

media_dim.toPandas()
```

#### Fact participation in media

```python
media_dim.createOrReplaceTempView("media_dim")

samuel_media_participation_fact = spark.sql(
"""
SELECT id AS media_id, genre_ids, first_air_date as release_date, popularity as media_popularity, vote_average as media_vote_average, vote_count as media_vote_count
FROM series_df

UNION

SELECT id as media_id, genre_ids, release_date, popularity as media_popularity, vote_average as media_vote_average, vote_count as media_vote_count
FROM movies_df
"""
)

samuel_media_participation_fact.createOrReplaceTempView("media_par")
date_dim.createOrReplaceTempView("date_dim")

samuel_media_participation_fact = spark.sql(
"""
SELECT mp.media_id, mp.genre_ids, mp.media_popularity, mp.media_vote_average, mp.media_vote_count, dd.date_id
FROM media_par mp
LEFT JOIN date_dim dd ON mp.release_date = dd.date
"""
)

samuel_media_participation_fact.toPandas()
```

```python
# The fact table should contain the FKs for the genres, not a list with genre_ids.

# exploding the genres list
samuel_media_participation_fact = samuel_media_participation_fact.withColumn("genre_id", f.explode_outer("genre_ids"))

# removing the genres list
samuel_media_participation_fact = samuel_media_participation_fact.drop("genre_ids")

samuel_media_participation_fact.toPandas()
```

### Saving the fact and dimensions

```python
genre_dim.write.parquet("../Gold/genres/dim_genre.parquet", mode="overwrite")

media_dim.write.parquet("../Gold/media/dim_media.parquet", mode="overwrite")

date_dim.write.parquet("../Gold/date/dim_date.parquet", mode="overwrite", partitionBy="date_year")

samuel_media_participation_fact.write.parquet("../Gold/media_participation/samuel_fact_media_participation.parquet", mode="overwrite")
```
